```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('divergence_heatmap.csv')
vals = np.array(df)
```

```python
fig,ax = plt.subplots(1,1,figsize=(13,13))
ax.imshow(vals, cmap = 'viridis')
# ax[1].imshow(vals, cmap = 'hsv')
# ax[2].imshow(vals, cmap = 'inferno')
fig.tight_layout()
# for ax in ax:
#     ax.xaxis.set_tick_params(labelcolor='none')
#     ax.yaxis.set_tick_params(labelcolor='none')
#     ax.axis('off') 
ax.xaxis.set_tick_params(labelcolor='none')
ax.yaxis.set_tick_params(labelcolor='none')
ax.axis('off')   
fig.savefig("chaos_heatmap.jpg", dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

def plot_bu_heatmap_clean(grid, filename="bu_chaos_heatmap.jpg"):
    # Define BU colors: white  scarlet  deep red
    bu_colors = ["#FFFFFF", "#CC0000", "#4B0000"]
    cmap_bu = LinearSegmentedColormap.from_list("BU", bu_colors)

    # Set up the figure: 3900x3900 pixels
    plt.figure(figsize=(13, 13), dpi=300)
    plt.imshow(grid, cmap=cmap_bu, origin='lower')

    # Remove all axes, ticks, labels, etc.
    plt.axis('off')
plot_bu_heatmap_clean(vals)
```
